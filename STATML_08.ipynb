# Question-1

**Below is an example dataset with 10 students (rows), 5 features, and a binary target label (Pass/Fail). The target label is 1 if the student is predicted to pass, and 0 if they are predicted to fail.**

```python
import pandas as pd
data = pd.read_csv('/content/csv.csv')
data.head()
```

```python
df=pd.DataFrame(data)
df
```

```python
x=df[['x1','x2','x3','x4','x5']]
y=df['pass/fail']#target
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=44)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

# Question-2

**Implement a logistic regression model to train above data, and find the accuracy.**

```python
import numpy as np
import pandas as pd
import matplotlib as plt
data=pd.read_csv('/archive (4).zip')
data
```

```python
gender = {"Male":1, "Female":0}
polyuria = {"Yes":1, "No":0}
polydipsia = {"Yes":1, "No":0}
sudden_weight_loss = {"Yes":1, "No":0}
weakness = {"Yes":1, "No":0}
polyphagia = {"Yes":1, "No":0}
genital_thrush = {"Yes":1, "No":0}
visual_blurring = {"Yes":1, "No":0}
itching = {"Yes":1, "No":0}
irritability = {"Yes":1, "No":0}
delayed_healing = {"Yes":1, "No":0}
partial_paresis = {"Yes":1, "No":0}
muscle_stiffness = {"Yes":1, "No":0}
alopecia = {"Yes":1, "No":0}
obesity = {"Yes":1, "No":0}
class_col = {"Positive":1, "Negative":0}

```

```python
data['Gender']=data['Gender'].map(gender)
data['Polyuria']=data['Polyuria'].map(polyuria)
data['Polydipsia']=data['Polydipsia'].map(polydipsia)
data['sudden weight loss']=data['sudden weight loss'].map(sudden_weight_loss)
data['weakness']=data['weakness'].map(weakness)
data['Polyphagia']=data['Polyphagia'].map(polyphagia)
data['Genital thrush']=data['Genital thrush'].map(genital_thrush)
data['visual blurring']=data['visual blurring'].map(visual_blurring)
data['Itching']=data['Itching'].map(itching)
data['Irritability']=data['Irritability'].map(irritability)
data['delayed healing']=data['delayed healing'].map(delayed_healing)
data['partial paresis']=data['partial paresis'].map(partial_paresis)
data['muscle stiffness']=data['muscle stiffness'].map(muscle_stiffness)
data['Alopecia']=data['Alopecia'].map(alopecia)
data['Obesity']=data['Obesity'].map(obesity)
data['class']=data['class'].map(class_col)
data
```

```python
print(data.dtypes)
```

```python
y = data['class']
x = data.drop(['class'], axis=1)
```

```python
nd = x-x.min()/x.max()-x.min()
nd
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=44)
x_train
x_test
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
mm=lr.fit(x_train,y_train)
```

```python
print(mm.intercept_)
print(mm.coef_)
```

```python
from sklearn.metrics import accuracy_score
```

```python
yp_train=lr.predict(x_train)
yp_test=lr.predict(x_test)
```

```python
print(accuracy_score(y_train,yp_train))
```

```python
print(accuracy_score(y_test,yp_test))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_train,yp_train))
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_train,yp_train))
```
